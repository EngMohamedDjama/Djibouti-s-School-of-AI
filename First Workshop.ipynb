## if,elif, else Statements

```python
# Condition must be satisfied in order to execute the following command else nothing will prompt

if 1 < 2:
    print('Yep!')
```

```python
if 1 > 2:
    print('Yep!')
```

```python
# Here we set a considition and set and outcome if that condition isn't satisfied then a generic output will prompt

if 1 < 2:
    print('first')
else:
    print('last')
```

```python
if 1 > 2:
    print('first')
else:
    print('last')
```

```python
# More condition but the same structure 

if 1 == 2:
    print('first')
elif 3 == 3:
    print('middle')
else:
    print('Last')
```

## for Loops

```python
# We instantiate a sequence first

seq = [1,2,3,4,5]
```

```python
for item in seq:
    print(item)
```

```python
for item in seq:
    print('there are this number of items')
```

```python
for x in seq:
    print(x+x)
```

## while Loops

```python
# Here we don't need to have a sequence but rather a condition to follow 

i = 1
while i < 5:
    print('i is: {}'.format(i))
    i = i+1
```

## list comprehension

```python
x = [1,2,3,4]
```

```python
out = []
for item in x:
    out.append(item**2)
print(out)
```
